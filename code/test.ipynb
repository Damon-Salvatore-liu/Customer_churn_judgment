```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
train_data = pd.read_csv('../data/train.csv')
test_data = pd.read_csv('../data/test_noLabel.csv')
```

```python
print('shape of train: ', train_data.shape)
print('shape of test: ', test_data.shape)
```

```python
train_X = train_data.drop(['Contract','PaymentMethod','Label'],axis=1)
```

```python
test_X = test_data.drop(['Contract','PaymentMethod'],axis=1)
```

```python
train_y = np.where(train_data['Label']=='Yes', 1, 0)
```

```python
train_X['DeviceProtection']=train_X['DeviceProtection'].str.split('').str[0].str.strip()
train_X['DeviceProtection']=np.where(train_X['DeviceProtection']=='No',0,1)
train_X['TVProgram']=train_X['TVProgram'].str.split('').str[0].str.strip()
train_X['TVProgram']=np.where(train_X['TVProgram']=='No',0,1)
train_X['Dependents']=np.where(train_X['Dependents']=='No',0,1)
train_X['InternetService']=np.where(train_X['InternetService']=='No',0,1)
train_X['MultipleLines']=np.where(train_X['MultipleLines']=='No',0,1)
train_X['Partner']=np.where(train_X['Partner']=='No',0,1)
train_X['PhoneService']=np.where(train_X['PhoneService']=='No',0,1)
train_X['gender']=np.where(train_X['gender']=='No',0,1)
```

```python
train_X
```

```python
knn = KNeighborsClassifier()
knn.fit(train_X, train_y)
```

```python
test_X['DeviceProtection']=test_X['DeviceProtection'].str.split('').str[0].str.strip()
test_X['DeviceProtection']=np.where(test_X['DeviceProtection']=='No',0,1)
test_X['TVProgram']=test_X['TVProgram'].str.split('').str[0].str.strip()
test_X['TVProgram']=np.where(test_X['TVProgram']=='No',0,1)
test_X['Dependents']=np.where(test_X['Dependents']=='No',0,1)
test_X['InternetService']=np.where(test_X['InternetService']=='No',0,1)
test_X['MultipleLines']=np.where(test_X['MultipleLines']=='No',0,1)
test_X['Partner']=np.where(test_X['Partner']=='No',0,1)
test_X['PhoneService']=np.where(test_X['PhoneService']=='No',0,1)
test_X['gender']=np.where(test_X['gender']=='No',0,1)
```

```python
y_pred = knn.predict(test_X)
```

```python
submit = test_X['ID']
```

```python
df = pd.DataFrame(columns=['ID','Label'])
```

```python
df['ID'], df['Label'] = test_X['ID'], y_pred
```

```python
df['Label']=np.where(df['Label']==0,'No','Yes')
```
